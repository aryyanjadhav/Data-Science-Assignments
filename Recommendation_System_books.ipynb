```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
df=pd.read_csv('book-2.csv',encoding='Latin1')
df
```

```python
df2=df.iloc[:,1:]
df2
```

```python
#sorting using user id
df2.sort_values(['User.ID'])
```

```python
#unique users
len(df2['User.ID'].unique())
```

```python
#unique books
len(df2['Book.Title'].unique())
```

```python
#pivotting the data
df3=df2.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
df3
```

```python
#assigning unique users as index
df3.index=df2['User.ID'].unique()
df3
```

```python
#replacing NaN values with 0
df3.fillna(0,inplace=True)
df3
```

```python
#calculating cosine similarity
usersim=1-pairwise_distances(df3.values,metric='cosine')
usersim
```

```python
#storing in dataframe
usersim2=pd.DataFrame(usersim)
usersim2
```

```python
#setting index and column names to user id
usersim2.index=df2['User.ID'].unique()
usersim2.columns=df2['User.ID'].unique()
usersim2
```

```python
#Making diagonal values as 0
np.fill_diagonal(usersim,0)
usersim2
```

```python
#finding similar users
usersim2.idxmax(axis=1)
```
