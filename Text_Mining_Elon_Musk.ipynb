```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import spacy
import string
from matplotlib.pyplot import imread
from wordcloud import WordCloud, STOPWORDS
%matplotlib inline
```

```python
df=pd.read_csv('Elon_musk.csv',encoding='Latin-1')
```

```python
df
```

```python
df.drop(['Unnamed: 0'],inplace=True,axis=1)
df
```

# **Preprocessing**

```python
#removing both leading and trailing characters
df=[Text.strip() for Text in df.Text]

#removing empty strings
df=[Text for Text in df if Text]

df[0:10]
```

```python
#joining all tweets into one text
df_text=' '.join(df)
df_text
```

```python
#removing usernames
from nltk.tokenize import TweetTokenizer
tz=TweetTokenizer(strip_handles=True)
df_tokens=tz.tokenize(df_text)
print(df_tokens)
```

```python
#rejoining
df_tokens_text=' '.join(df_tokens)
df_tokens_text
```

```python
#removing punctuations
nop_text=df_tokens_text.translate(str.maketrans('','',string.punctuation))
nop_text
```

```python
#removing urls
import re
nou_text=re.sub(r'http\S+','',nop_text)
nou_text
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
txt_tokens=word_tokenize(nou_text)
txt_tokens
```

```python
#tokenization
nltk.download('stopwords')
```

```python
len(txt_tokens)
```

```python
#removing stopwords
from nltk.corpus import stopwords
stp=stopwords.words('english')
sw=['\x92','rt','ye','yeah','haha','Yes','U0001F923','I']
stp.extend(sw)

no_stop_tokens=[word for word in txt_tokens if not word in stp]
print(no_stop_tokens)
```

```python
#normalization
lw=[Text.lower() for Text in no_stop_tokens]
print(lw[100:200])
```

```python
#stemming (not so accurate)
from nltk.stem import PorterStemmer
ps=PorterStemmer()
st=[ps.stem(word) for word in lw]
print(st[100:200])
```

```python
#lemmatization
nlp=spacy.load('en_core_web_sm')
doc=nlp(' '.join(lw))
doc
```

```python
lemmas=[token.lemma_ for token in doc]
lemmas
```

```python
clean=' '.join(lemmas)
clean
```

# **Named Entity Recognition(NER)**

```python
#POS tagging
nlp=spacy.load('en_core_web_sm')
ob=clean
db=nlp(ob)
spacy.displacy.render(db,style='ent',jupyter=True)
```

```python
for token in db[100:200]:
  print(token,token.pos_)
```

```python
#filtering nouns and verbs
nv=[token.text for token in db if token.pos_ in ('NOUN','VERB')]
print(nv[100:200])
```

```python
#counting nouns and verbs
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()

X=cv.fit_transform(nv)
sum_words=X.sum(axis=0)

words_freq=[(word,sum_words[0,idx]) for word,idx in cv.vocabulary_.items()]
words_freq=sorted(words_freq, key=lambda x: x[1], reverse=True)

wd_df=pd.DataFrame(words_freq)
wd_df.columns=['word','count']
wd_df[0:10]
```

```python
#visualization
wd_df[0:10].plot.bar(x='word',figsize=(12,8),title='Top 10 nouns and verbs');
```

# **Sentiment Analysis**

```python
from nltk import tokenize
sentences=tokenize.sent_tokenize(' '.join(df))
sentences
```

```python
sentdf=pd.DataFrame(sentences,columns=['sentence'])
sentdf
```

```python
afn=pd.read_csv('Afinn.csv')
afn
```

```python
affinity_scores=afn.set_index('word')['value'].to_dict()
affinity_scores
```

```python
# Custom function: score each word in a sentence in lemmatised form, but calculate the score for the whole original sentence
nlp=spacy.load('en_core_web_sm')
sentiment_lexicon=affinity_scores

def calculate_sentiment(text:str=None):
    sent_score=0
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score+=sentiment_lexicon.get(word.lemma_,0)
    return sent_score
```

```python
#manual testing
calculate_sentiment(text='great')
```

```python
#manual testing
calculate_sentiment(text='awful')
```

```python
#Calculating sentiment value for each sentence
sentdf['sentiment_value']=sentdf['sentence'].apply(calculate_sentiment)
sentdf['sentiment_value']
```

```python
#how many words are there in a sentence?
sentdf['word_count']=sentdf['sentence'].str.split().apply(len)
sentdf['word_count']
```

```python
sentdf.sort_values(by='sentiment_value')
```

```python
#negative sentiment score
sentdf[sentdf['sentiment_value']<=0]
```

```python
#positive sentiment score
sentdf[sentdf['sentiment_value']>0]
```

```python
sentdf['index']=range(0,len(sentdf))
sentdf
```
