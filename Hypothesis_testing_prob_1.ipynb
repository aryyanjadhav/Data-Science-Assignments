```python
'''
assuming null hypothesis Ho -> μ1 = μ2 (There is no difference in diameters of cutlets between two units)
alternate hypothesis Ha -> μ1 ≠ μ2 (There is significant difference in diameters of cutlets between two units) 2 Sample 2 Tail test applicable
'''
```

```python
import pandas as pd
from scipy import stats
from scipy.stats import norm
```

```python
data=pd.read_csv('Cutlets.csv')
data
```

```python
Unit_A=pd.Series(data.iloc[:,0])
Unit_B=pd.Series(data.iloc[:,1])
```

```python
# 2-sample 2-tail ttest:   stats.ttest_ind(array1,array2)     # ind -> independent samples
p_value=stats.ttest_ind(Unit_A,Unit_B)
p_value
```

```python
p_value[1]
```

```python
'''
on comparing p_value with α = 0.05 (significance level 5%)
p_value>α, hence failed to reject null hypothesis (Ho), accept Ho.
'''
```
