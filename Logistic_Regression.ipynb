```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,roc_auc_score,roc_curve
```

```python
bank=pd.read_csv('bank-full.csv',delimiter=';')
bank
```

# **EDA**

```python
bank.info()
```

```python
#One-hot encoding of categorical variables
data=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
data
```

```python
#displaying all columns
pd.set_option('display.max.columns',None)
data
```

```python
data.info()
```

```python
#Custom binary encoding
data['default']=np.where(data['default'].str.contains('yes'),1,0)
data['housing']=np.where(data['housing'].str.contains('yes'),1,0)
data['loan']=np.where(data['loan'].str.contains('yes'),1,0)
data['y']=np.where(data['y'].str.contains('yes'),1,0)
```

```python
data
```

```python
data['month'].value_counts()
```

```python
month1={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
data=data.replace(month1)
data
```

```python
data.info()
```

# **MODEL**

```python
#dividing data into input & output
x=pd.concat([data.iloc[:,0:11],data.iloc[:,12:]],axis=1)
y=data.iloc[:,11]
```

```python
#LR model
model=LogisticRegression()
model.fit(x,y)
```

# **PREDICTIONS**

```python
ypred=model.predict(x)
ypred
```

```python
ypred_df=pd.DataFrame({'Actual_y':y,'Predicted_y':ypred})
ypred_df
```

# **Model Accuracy**

```python
#Confusion Matrix
cm=confusion_matrix(y,ypred)
cm
```

```python
#accuracy=(a+d)/(a+b+c+d)
(39032+1310)/(39032+890+3979+1310)
```

# **Model accuracy is 89.23%**

```python
model.predict_proba(x)[:,1]
```

```python
#ROC curve & AUC accuracy
fpr,tpr,threshold=roc_curve(y,model.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='purple')

auc=roc_auc_score(y,ypred)

plt.plot(fpr,tpr,color='red')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```

**AUC accuracy is 61.27% , this score here is actually not good.**
